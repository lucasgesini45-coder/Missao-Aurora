```python
import random
import time
import pandas as pd

#------------------------------------------------------------------------------------------
# Fase Inicial Da Missao.
print("==================================\n       MISSAO AURORA SIGER\n==================================")
resposta = input("Deseja iniciar o processo para lançamento do Foguete? sim/nao ").strip().lower()

if resposta == "sim":
    print("Iniciando processo de verificação...\nAveriguando dados...")
    time.sleep(1.5)

    etapas = [
        "Temperatura Externa",
        "Temperatura Interna",
        "Energia do tanque",
        "Pressão do tanque",
        "Integridade Estrutural",
        "Módulos Críticos"
    ]

    for etapa in etapas:
        print(f"Verificando {etapa}...")
        time.sleep(0.5)

elif resposta == "nao":
    print("O lançamento foi cancelado!")
    exit()
else:
    print("Resposta inválida!")
    exit()

#-------------------------------------------------------------------------------------------
# Geração de dados aleatórios mais realistas

temp_externa1 = random.randint(110, 120)
temp_interna2 = random.randint(20, 30)
energia3 = random.randint(3, 8)
pressao_tanque4 = random.randint(85, 105)
integridade_estrutural5 = random.choice([0, 1])
modulos_criticos6 = random.choice([0, 1])

#-------------------------------------------------------------------------------------------
# Status

status_temp_ext = "OK" if temp_externa1 >= 115 else "N/G"
status_temp_int = "OK" if temp_interna2 >= 25 else "N/G"
status_energia = "OK" if energia3 >= 6 else "N/G"
status_pressao = "OK" if pressao_tanque4 >= 99 else "N/G"
status_integridade = "OK" if integridade_estrutural5 == 1 else "N/G"
status_modulos = "OK" if modulos_criticos6 == 1 else "N/G"

#-------------------------------------------------------------------------------------------
# DataFrame

data = {
    'Parâmetro': ['Temp Externa', 'Temp Interna', 'Energia', 'Pressão Tanque', 'Integridade Estrutural', 'Módulos Críticos'],
    'Valor': [temp_externa1, temp_interna2, energia3, pressao_tanque4, integridade_estrutural5, modulos_criticos6],
    'Status': [status_temp_ext, status_temp_int, status_energia, status_pressao, status_integridade, status_modulos]
}

df_telemetria = pd.DataFrame(data)

print("\n=================================================")
print("           DADOS DO FOGUETE 🚀")
print("=================================================")
print(df_telemetria)

#-------------------------------------------------------------------------------------------
# Decisão final

if all(status == "OK" for status in data['Status']):
    print("\n🚀 Foguete autorizado para a decolagem!")
    print("Iniciando decolagem...")
else:
    print("\n❌ Decolagem não autorizada!")
    print("Status da nave fora dos parâmetros.")
```
