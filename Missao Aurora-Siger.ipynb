```python
#importaçao das classes
import random
import time
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

#------------------------------------------------------------------------------------------
# Fase Inicial Da Missao
print("==================================")
print("       MISSAO AURORA SIGER")
print("==================================")

resposta = input("Deseja iniciar o processo para lançamento do Foguete? (sim/nao): ").strip().lower()

if resposta == "nao":
    print("❌ O lançamento foi cancelado!")
    exit()
elif resposta != "sim":
    print("❌ Resposta invalida!")
    exit()

print(f"Iniciando verificações...")
{time.sleep(0.5)}
print(f"Verificando temperatura externa...")
{time.sleep(0.5)}
print(f"Verificando temperatura interna...")
{time.sleep(0.5)}
print(f"Verificando energia...")
{time.sleep(0.5)}
print(f"verificando integridade...  ")
{time.sleep(0.5)}
print(f"verificando modulos...")
{time.sleep(0.5)}
#------------------------------------------------------------------------------------------
# Declaraçao dos dados

temp_externa = random.choice([15, 30])
temp_interna = random.choice([15, 40])
energia = random.choice([28, 32])
pressao = random.choice([2, 5])
integridade = random.choice([95, 100])
modulos = random.choice([0, 1])

#------------------------------------------------------------------------------------------
# Tabela de telemetria

df = pd.DataFrame({
    "Parametro": ["Temp Externa", "Temp Interna", "Energia", "Pressao", "Integridade", "Modulos"],
    "Valor": [temp_externa, temp_interna, energia, pressao, integridade, modulos]
})

print("\n📊 TELEMETRIA DO FOGUETE")
print(df)

#------------------------------------------------------------------------------------------
# ⚡ ANÁLISE ENERGÉTICA

print("\n⚡ ANÁLISE ENERGÉTICA")

capacidade_total = random.randint(800, 1200)
carga_percentual = random.randint(60, 100)
consumo = random.randint(200, 400)
perdas = random.uniform(0.05, 0.15)

energia_disp = capacidade_total * (carga_percentual / 100) * (1 - perdas)
autonomia = energia_disp / consumo

status_energia = "OK" if autonomia >= 1 else "CRITICO"

df_energia = pd.DataFrame({
    "Parâmetro": [
        "Capacidade Total (kWh)",
        "Carga Atual (%)",
        "Perdas (%)",
        "Energia Disponível (kWh)",
        "Consumo Decolagem (kWh)",
        "Autonomia (ciclos)"
    ],
    "Valor": [
        capacidade_total,
        carga_percentual,
        round(perdas * 100, 2),
        round(energia_disp, 2),
        consumo,
        round(autonomia, 2)
    ]
})

print(df_energia)

if status_energia == "OK":
    print("✅ Energia suficiente")
else:
    print("❌ Energia insuficiente")

#------------------------------------------------------------------------------------------
#IA SCIKIT-LEARN

dados_treino = [
    [30, 40, 32, 5, 100, 1, 1],
    [15, 15, 28, 2, 95, 0, 0],
    [30, 40, 32, 5, 100, 1, 1],
    [15, 40, 28, 2, 95, 1, 0]
]

colunas = ["temp_ext", "temp_int", "energia", "pressao", "integridade", "modulos", "lancar"]
df_treino = pd.DataFrame(dados_treino, columns=colunas)

X = df_treino.drop("lancar", axis=1)
y = df_treino["lancar"]

modelo = DecisionTreeClassifier()
modelo.fit(X, y)

# Previsão da IA
entrada = [[temp_externa, temp_interna, energia, pressao, integridade, modulos]]
previsao = modelo.predict(entrada)

#------------------------------------------------------------------------------------------
# DECISÃO FINAL

print("DECISÃO FINAL")

if previsao[0] == 1 and status_energia == "OK":
    print("🚀A DECOLAGEM FOI AUTORIZADA!")
else:
    print("❌ PARAMETROS DE DECOLAGEM NAO ESTA FUNCIONDO CORRETAMENTE DECOLAGEM NEGADA!")
```
